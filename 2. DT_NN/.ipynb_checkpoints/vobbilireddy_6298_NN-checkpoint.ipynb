```python
# while selecting best 3 attributes for training,validating and testing your model. [2 points]
# with the help of correlation graph and scatter plot we were able to find the top three attributes
# they are # preg,plas,age
```

```python
#Standardize the features  
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
```

```python
# select the top 3 features and drop the remaining features dataX used as input features and dataY used as Target
dataX =  pd.DataFrame(sc_X.fit_transform(dataset.drop(['skin','Pres','test','mass','pedi','class'],axis = 1),),
        columns=['Preg','Plas','age'])
dataY = dataset["class"]
print(dataX.columns)
```

# Find the best K (we are using elbow method to find the best k , where it automatically determines optimal k value for the dataset)

```python
from sklearn.cluster import KMeans
wcss = [] # list of Within-Cluster Sum of Squares values are stored inthe list created
for var in range(1, 11):  # the range of clusters i.e total 11 Clusters we are going to check 
    k_val = KMeans(n_clusters = var, init = 'k-means++', random_state = 2022)
    k_val.fit(dataset) # fit the datavalues in the cluster created using KMeans function
    wcss.append(k_val.inertia_) 
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('wcss')  
plt.show()
```

# Split your dataset # 60% for training set, 20% for Validation set and 20% for testing set the classifier. Use random state as 2022 [2 points]

```python
x_train, x_test, y_train, y_test = train_test_split(dataX, dataY,test_size=0.2, random_state = 2022)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state= 2022) 
print(x_train.shape, x_test.shape, x_val.shape)
print(y_train.shape, y_test.shape, y_val.shape)
```

# Use minkowski distance.

```python
from sklearn.neighbors import KNeighborsClassifier
knn_minkowski = KNeighborsClassifier(n_neighbors=19, n_jobs=-1,metric='minkowski')
knn_minkowski.fit(x_train,y_train)
print("Validation test scores are:",knn_minkowski.score(x_val,y_val))
print("Testing scores are:",knn_minkowski.score(x_test,y_test))
```

# Test the classifier with three different numbers for neighbors and record the results for training, validation and testing. Show the best one.[3 points]

```python
#We have initialized three empty lists and we are fitting the accuracy for 3 different values of k as questioned.
#knn1,2,3 contains the class definedd to it with the parameters, then we fit it and append the scores to the empty list as initilaized. 
```

```python
#Setup a knn classifier with k neighbors

result_val=[]
result_test=[]
knn1 = KNeighborsClassifier(n_neighbors=19, n_jobs=-1,metric='minkowski')
knn1.fit(x_train,y_train)
result_val.append(knn1.score(x_val,y_val))
result_test.append(knn1.score(x_test,y_test))
print("KNN Score for validation data",knn1.score(x_val,y_val))
print("KNN Score for training data",knn1.score(x_test,y_test))
```

```python
#Setup a knn classifier with k neighbors
knn2 = KNeighborsClassifier(n_neighbors=25, n_jobs=-1,metric='minkowski')
knn2.fit(x_train,y_train)
result_val.append(knn2.score(x_val,y_val))
result_test.append(knn2.score(x_test,y_test))
print("KNN Score for validation data",knn2.score(x_val,y_val))
print("KNN Score for training data",knn2.score(x_test,y_test))
```

```python
#Setup a knn classifier with k neighbors
knn3 = KNeighborsClassifier(n_neighbors=35, n_jobs=-1,metric='minkowski')
knn3.fit(x_train,y_train)
result_val.append(knn3.score(x_val,y_val))
result_test.append(knn3.score(x_test,y_test))
print("KNN Score for validation data",knn3.score(x_val,y_val))
print("KNN Score for training data",knn3.score(x_test,y_test))
```

```python
best_knn_val= max(result_val)
count= 0
for i in (result_val):
    count+=1
    #print(count)
    if i == best_knn_val:
        print("Max score:{} and cluster:{}  ".format(best_knn_val,count))
```

```python
## score that comes from testing on the same datapoints that were used for training
best_knn_train = max(result_test)
count= 0
for i in (result_test):
    count+=1
    #print(count)
    if i == best_knn_train:
        print("Max score:{} and cluster:{}  ".format(best_knn_train,count))
```

###  Plotting the train and test results after fitting the data into the model using minkowski distance

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(range(1,4),result_test,marker='*',label='Train results')
p = sns.lineplot(range(1,4),result_val,marker='o',label='Test results')
```

# Calculate and print the confusion matrix, and the classification Report (includes:precision, recall, f1-score, and support) for all three different numbers.

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
y_pred1_test = knn1.predict(x_test)
y_pred2_test = knn2.predict(x_test)
y_pred3_test = knn3.predict(x_test)
y_pred1_train = knn1.predict(x_train)
y_pred2_train = knn2.predict(x_train)
y_pred3_train= knn3.predict(x_train)
cnf_matrix1_test = confusion_matrix(y_test, y_pred1_test)
cnf_matrix2_test = confusion_matrix(y_test, y_pred2_test)
cnf_matrix3_test = confusion_matrix(y_test, y_pred3_test)
cnf_matrix1_train = confusion_matrix(y_train, y_pred1_train)
cnf_matrix2_train = confusion_matrix(y_train, y_pred2_train)
cnf_matrix3_train= confusion_matrix(y_train, y_pred3_train)
#Prediction
print("******Confusion matrix & Classification report on test data********************************")
print("Confusion matrix1 :\n",cnf_matrix1_test)
print("classification report on test data for neighbour 1:\n",classification_report(y_pred1_test,y_test))
print("Confusion matrix2:\n",cnf_matrix2_test)
print("classification report on test data for neighbour 2:\n",classification_report(y_pred2_test,y_test))
print("Confusion matrix3:\n",cnf_matrix3_test)
print("classification report on test data for neighbour 3:\n",classification_report(y_pred3_test,y_test))
print("****** Confusion matrix report on train  data********************************")
print("Confusion matrix1:\n",cnf_matrix1_train)
print("classification report on train data for neighbour 1:\n",classification_report(y_pred1_train,y_train))
print("Confusion matrix2:\n",cnf_matrix2_train)
print("classification report on train data for neighbour 2:\n",classification_report(y_pred2_train,y_train))
print("Confusion matrix3:\n",cnf_matrix3_train)
print("classification report on train data for neighbour 3:\n",classification_report(y_pred3_train,y_train))
```
