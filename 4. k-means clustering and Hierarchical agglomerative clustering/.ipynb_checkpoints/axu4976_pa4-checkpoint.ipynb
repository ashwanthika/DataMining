## TASK 1: K-Means Clustering

### Task 1-a: Determine â€œkâ€ value from the elbow method

In this task, you will be using the elbow method to determine the optimal number of clusters for k-means clustering.

We need some way to determine whether we are using the right number of clusters when using k-means clustering. One method to validate the number of clusters is the elbow method. 

The idea of the elbow method is to run k-means clustering on the dataset for a range of values of k (k will be from 1 to 10 in this task), and for each value of k calculate the sum of squared errors (SSE). Then, plot a line chart of the SSE for each value of k. If the line chart looks like an arm, then the "elbow" on the arm is the value of k that is the best. The idea is that we want a small SSE, but that the SSE tends to decrease toward 0 as we increase k (the SSE is 0 when k is equal to the number of data points in the dataset, because then each data point is a cluster, and there is no error between it and the center of its cluster). So our goal is to choose a small value of k that still has a low SSE, and the elbow usually represents where we start to have diminishing returns by increasing k.

For this task, you need to perform the elbow method for k from 1 to 10 and plot a line chart of the SSE for each value of k, and determine the best k (the number of clusters). Note that you need to use the whole dataset in this task and you need to print your decision for k.

```python
#########################begin code for Task 1-a
#########################begin code for Task 1-a
from sklearn.cluster import KMeans
sum_of_squre_err = [] #creating an empty list to append the sum of errors for each k value later. 
for var in range(1, 11): #range of k value is from 1-10. thus, we give 1-11. 11 is ignored 
    k_val = KMeans(n_clusters = var, init = 'k-means++', random_state = 2022) #creating a k value object with kmeans function
    k_val.fit(college_df) #we fit our dataset to the previous created object 
    sum_of_squre_err.append(k_val.inertia_) #claculating the sum of errors for each k value  
plt.plot(range(1, 11), sum_of_squre_err, marker='X', markeredgecolor='pink', markerfacecolor='black') #plotting the elbow curve for the given k values as x axis and sum of square errors on y axis
plt.title('Elbow Method')
plt.xlabel('k value')
plt.ylabel('sum_of_squre_err')  
plt.show()
print("Elbow method helps us to find the optimal k value. post the suggested k value, the graph will decrease less drastically. We choose our k value to be 3. If you see the graph, after k value 3, the sum_of_squared value decreases less")
```

### Task 1-b: Visualization for K-Means Clustering

In this task, you will be performing k-means clustering for k=2 and visualize the predicted training samples and actual training samples on scatter plots. Use 70% of the dataset for training and 30% of the dataset for testing. Perform kmeans for clustering samples in your training set. 

Use two subplots for visualizing the predicted training samples and actual training samples on two scatter plots.

Since your dataset has multiple features(dimensions), you won't be able to plot your data on a scatter plot. Thus, youâ€™re going to visualize your data with the help of one of the Dimensionality Reduction techniques, namely Principal Component Analysis (PCA). The idea in PCA is to find a linear combination of the two variables that contains most of the information. This new variable or â€œprincipal componentâ€ can replace the two original variables. You can easily apply PCA to your data with the help of scikit-learn.



```python
###################begin code for Task 1-b-1: Split the dataset 70% for training and 30% for testing
# random state = 2022
### Important!!!
tr_x = college_df.sample(frac=0.7, random_state=2022) #tr_x represenst train x
tt_x = college_df.drop(tr_x.index) #tt_x represenst test x
tr_y = tr_x.pop('Grad_Rate')  #tr_y represenst train y
tt_y = tt_x.pop('Grad_Rate') #tt_y represenst test y
#creating 70%train dataset and 30% test dataset follwed by popping the train and test predictor variable in the tr_y and tt_y variables
###################end code for Task 1-b-1
```

```python
###################begin code for Task 1-b-2: Visualize the predicted training labels vs actual training labels

# Import PCA
from sklearn.decomposition import PCA


# Create the KMeans model
model_knn = KMeans(n_clusters = 2, random_state=2022) #kmeans object creation with 2 clsuters and random state 2022
model_knn.fit(tr_x) #fitting our model ot the knn
# Compute cluster centers and predict cluster index for each sample 
cc_index = model_knn.predict(tr_x) #predicting our train data and finding out the cluster index 
# Model and fit the data to the PCA model
pca_model= PCA(n_components=2) #creating a pca model obj. using pca we minimize dimensions of the variable based on variables that have high variance
X_train_pca = pca_model.fit_transform(tr_x) #fitting our train datasplit to the model

# Visualize the predicted training labels vs actual training labels. 
### scatter(x, y, your_data)

x = X_train_pca[:, 0] #will have all x coordinate values
y = X_train_pca[:, 1] #will have all y coordinate values
Y_train_pca = model_knn.fit_predict(X_train_pca) #fitting the model and getting the predicted clusters using knn 

#plotting the graph for actual class labels of training set
plt.show()
plt.figure(figsize=(9,9))
plt.title("actual training data labels")
plt.scatter(x,y, c=tr_y)

#plotting the graph for predicted class labels using knn
plt.figure(figsize=(9,9))
plt.title("predicted training data labels")
plt.scatter(x,y, c=cc_index)
plt.scatter(model_knn.cluster_centers_[:, 0],
            model_knn.cluster_centers_[:, 1],
             marker='x',
            c='green',
            s=300,label='centroids')


###################end code for Task 1-b-2
```

Now, you need to visualize the predicted testing labels versus actual testing labels. Use the trained model in previous step.

```python
###################begin code for Task 1-b-3: Visualize the predicted testing labels vs actual testing labels

model_knn_test = KMeans(n_clusters = 2, random_state=2022).fit(tt_x) #kmeans object creation with 2 clsuters and random state 2022

# predict cluster index for each sample 
cc_index_test = model_knn_test.predict(tt_x)  #predicting our test data and finding out the cluster index 
# Model and fit the data to the PCA model
pca_model_test= PCA(n_components=2) #creating a pca model obj. using pca we minimize dimensions of the variable based on variables that have high variance
X_test_pca = pca_model_test.fit_transform(tt_x) #fitting our test datasplit to the model


# Visualize the predicted testing labels vs actual testing labels. 
### scatter(x, y, your_data)
x = X_test_pca[:, 0] #will have all x coordinate values
y = X_test_pca[:, 1] #will have all y coordinate values
Y_test_pca = model_knn_test.fit_predict(X_test_pca) #fitting the model and getting the predicted clusters using knn 

#plotting the graph for actual class labels of test set
plt.show()
plt.figure(figsize=(9,9))
plt.title("actual test data labels")
plt.scatter(x,y, c=tt_y)
#plotting the graph for predicted class labels using knn
plt.figure(figsize=(9,9))
plt.title("predicted test data labels")
plt.scatter(x,y, c=cc_index_test)
plt.scatter(model_knn_test.cluster_centers_[:, 0],
            model_knn_test.cluster_centers_[:, 1],
             marker='x',
            c='green',
            s=300,label='centroids')



###################end code for Task 1-b-3
```

# TASK 2: Hierarchical Agglomerative  Clustering

## Task 2-a: Find the best Hierarchical Agglomerative Clustering Model

In this task, you will be performing Hierarchical Agglomerative clustering with different linkage methods (complete and average) and different similarity measures (cosine, euclidean, and manhattan) in order to find the best pair of linkage method and similarity measure. Use F1 score for evaluation and take n_clusters = 2.


```python
###################begin code for Task 2-a: Print out a confusion matrix
# Import AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
# Import pairwise_distances for calculating pairwise distance matrix
from sklearn.metrics.pairwise import pairwise_distances
# Import f1_score
from sklearn.metrics import f1_score



## Calculate pairwise distance matrix for X_train
pdm_train = tr_x

## Model and fit the training data to the AgglomerativeClustering model
## complete linkage + cosine
c_dst = pairwise_distances(pdm_train, metric='cosine')
h_cos = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='complete')
predicted_h_cos = h_cos.fit_predict(c_dst)
f1_score_cosine = f1_score(tr_y, predicted_h_cos)

## Model and fit the training data to the AgglomerativeClustering model
## complete linkage + euclidean
eu_dst = pairwise_distances(pdm_train, metric='euclidean')
h_eu = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='complete')
predicted_h_eu = h_eu.fit_predict(eu_dst)
f1_score_euclidean = f1_score(tr_y, predicted_h_eu)

## Model and fit the training data to the AgglomerativeClustering model
## complete linkage + manhattan
m_dst = pairwise_distances(pdm_train, metric='manhattan')
hierarchical_manhattan = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='complete')
pred_h_man = hierarchical_manhattan.fit_predict(m_dst)
f1_score_manhattan = f1_score(tr_y, pred_h_man)

## Model and fit the training data to the AgglomerativeClustering model
## average linkage + cosine
c_dst_avg = pairwise_distances(pdm_train, metric='cosine')
h_cos_avg = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='average')
predicted_h_cos_avg = h_cos_avg.fit_predict(c_dst_avg)
f1_score_cosine_avg = f1_score(tr_y, predicted_h_cos_avg)


## Model and fit the training data to the AgglomerativeClustering model
## average linkage + euclidean
eu_dst_avg = pairwise_distances(pdm_train, metric='euclidean')
h_eu_avg = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='average')
predicted_h_eu_avg = h_eu_avg.fit_predict(eu_dst_avg)
f1_score_euclidean_avg = f1_score(tr_y, predicted_h_eu_avg)


## Model and fit the training data to the AgglomerativeClustering model
## average linkage + manhattan
m_dst_avg = pairwise_distances(pdm_train, metric='manhattan')
h_man_avg = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='average')
pred_h_man_avg = h_man_avg.fit_predict(m_dst_avg)
f1_score_manhattan_avg = f1_score(tr_y, pred_h_man_avg)

print("F1-score for complete linkage + cosine", f1_score_cosine)
print("F1-score for complete linkage + euclidean", f1_score_euclidean)
print("F1-score for complete linkage + manhattan", f1_score_manhattan)
print("F1-score for average linkage + cosine", f1_score_cosine_avg )
print("F1-score for average linkage + euclidean", f1_score_euclidean_avg)
print("F1-score for average linkage + manhattan", f1_score_manhattan_avg)
#Agglomerative clustering uses a bottom-up approach, wherein each data point starts in its own cluster. These clusters are then joined, by taking the two most similar clusters together and merging them.
#In all the above functions, AgglomerativeClustering function has parameters, n_clusters means the number of clusters to find, affinity is used to find the distance, If â€œprecomputedâ€, a distance matrix is needed as input for the fit method. , linkage= The linkage criterion determines which distance to use
print("The model that is performing better is the one that has used average linkage with eucledian distance. It gives us an f1 score of 0.7867298578199051")
###################end code for Task 2-a
```
