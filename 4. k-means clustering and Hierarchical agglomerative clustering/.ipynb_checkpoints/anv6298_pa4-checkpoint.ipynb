## TASK 1: K-Means Clustering

### Task 1-a: Determine “k” value from the elbow method

In this task, you will be using the elbow method to determine the optimal number of clusters for k-means clustering.

We need some way to determine whether we are using the right number of clusters when using k-means clustering. One method to validate the number of clusters is the elbow method. 

The idea of the elbow method is to run k-means clustering on the dataset for a range of values of k (k will be from 1 to 10 in this task), and for each value of k calculate the sum of squared errors (SSE). Then, plot a line chart of the SSE for each value of k. If the line chart looks like an arm, then the "elbow" on the arm is the value of k that is the best. The idea is that we want a small SSE, but that the SSE tends to decrease toward 0 as we increase k (the SSE is 0 when k is equal to the number of data points in the dataset, because then each data point is a cluster, and there is no error between it and the center of its cluster). So our goal is to choose a small value of k that still has a low SSE, and the elbow usually represents where we start to have diminishing returns by increasing k.

For this task, you need to perform the elbow method for k from 1 to 10 and plot a line chart of the SSE for each value of k, and determine the best k (the number of clusters). Note that you need to use the whole dataset in this task and you need to print your decision for k.

```python
#########################begin code for Task 1-a
sse = [] #creating a list to store the sum of squared errord 
for var in range(1, 11):  
    k_val = KMeans(n_clusters = var, init = 'k-means++', random_state = 2022) #calling the default kmeans function for 10 values 1-10 as mentioned in the question
    k_val.fit(dataset)
    sse.append(k_val.inertia_) #Sum of squared distances of samples to their closest cluster center is computed by using intertia_
fig, ax = plt.subplots(figsize=(8, 6))
plt.plot(range(1, 11), sse, marker='x', markeredgecolor='black', markerfacecolor='black')
plt.title('SSE vs No. Of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared errors')  
plt.show()
print("--> Elbow graph is plotted below with x asis as the number of clusters and y axis as Sum of squared errors\n")
print("--> Elbow method is used to find the best K value for the given set of K values. In our case, k values range from 1-10. We choose the elbow point as the optimal value for k. post the elbow point, all values will generally decrease slowly/ in a linear manner. This is used as an indication to find out the k value. In the graph, we can propose the best k value as 3.")
```

### Task 1-b: Visualization for K-Means Clustering

In this task, you will be performing k-means clustering for k=2 and visualize the predicted training samples and actual training samples on scatter plots. Use 70% of the dataset for training and 30% of the dataset for testing. Perform kmeans for clustering samples in your training set. 

Use two subplots for visualizing the predicted training samples and actual training samples on two scatter plots.

Since your dataset has multiple features(dimensions), you won't be able to plot your data on a scatter plot. Thus, you’re going to visualize your data with the help of one of the Dimensionality Reduction techniques, namely Principal Component Analysis (PCA). The idea in PCA is to find a linear combination of the two variables that contains most of the information. This new variable or “principal component” can replace the two original variables. You can easily apply PCA to your data with the help of scikit-learn.



```python
###################begin code for Task 1-b-1: Split the dataset 70% for training and 30% for testing
# random state = 2022
### Important!!!
train_data_x = dataset.sample(frac=0.7, random_state=2022) #splitting the train data as 70%, we assign the random state to be 2022
test_data_x = dataset.drop(train_data_x.index) # rest 30% is assigned to the test data
train_data_y = train_data_x.pop('Class') #predictor variable of train data is assigned to train data y
test_data_y = test_data_x.pop('Class') #predictor variable of test data is assigned to train data y

###################end code for Task 1-b-1
```

```python
###################begin code for Task 1-b-2: Visualize the predicted training labels vs actual training labels

# Import PCA
from sklearn.decomposition import PCA

# Create the KMeans model
kmeans = KMeans(n_clusters = 2, random_state=2022) #the default Kmeans method is called, kmeans object is created. We aassume the number of clusters to be 2, with a random state of 2022
kmeans.fit(train_data_x) # fitting our train data to the kmeans obj
# Compute cluster centers and predict cluster index for each sample 
cluster_idx = kmeans.predict(train_data_x)
# Model and fit the data to the PCA model
principal_component= PCA(n_components=2)
X_train_pca = principal_component.fit_transform(train_data_x)

# Visualize the predicted training labels vs actual training labels. 
### scatter(x, y, your_data)

x = X_train_pca[:, 0]
y = X_train_pca[:, 1]
Y_train_pca = kmeans.fit_predict(X_train_pca)

Color_1 = ['black','green'] 
Color_2 = ['pink', 'yellow']
plt.figure(figsize=(10,8)) #plotting the graph for the train data that has output varibale 
plt.title("Predicted training labels ")
plt.scatter(x,y, c=cluster_idx, cmap=matplotlib.colors.ListedColormap(Color_1))
plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
             marker='x',
            c='blue',
            s=300,label='centroids')

plt.show()
plt.figure(figsize=(10,8)) #plotting the graph for which we have dropped the output variable
plt.title("Actual training labels")
plt.scatter(x,y, c=train_data_y, cmap=matplotlib.colors.ListedColormap(Color_2))
###################end code for Task 1-b-2
```

# TASK 2: Hierarchical Agglomerative  Clustering

## Task 2-a: Find the best Hierarchical Agglomerative Clustering Model

In this task, you will be performing Hierarchical Agglomerative clustering with different linkage methods (complete and average) and different similarity measures (cosine, euclidean, and manhattan) in order to find the best pair of linkage method and similarity measure. Use F1 score for evaluation and take n_clusters = 2.


```python
###################begin code for Task 2-a: Print out a confusion matrix
# Import AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
# Import pairwise_distances for calculating pairwise distance matrix
from sklearn.metrics.pairwise import pairwise_distances
# Import f1_score
from sklearn.metrics import f1_score

## Calculate pairwise distance matrix for X_train
pdm_train = train_data_x

## Model and fit the training data to the AgglomerativeClustering model
## complete linkage + cosine
cosine_dst = pairwise_distances(pdm_train, metric='cosine')
hierarchical_cosine = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='complete')
predicted_hierarchical_cosine = hierarchical_cosine.fit_predict(cosine_dst)
f1_score_cosine = f1_score(train_data_y, predicted_hierarchical_cosine)

## Model and fit the training data to the AgglomerativeClustering model
## complete linkage + euclidean
euclidean_dst = pairwise_distances(pdm_train, metric='euclidean')
hierarchical_euclidean = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='complete')
predicted_hierarchical_euclidean = hierarchical_euclidean.fit_predict(euclidean_dst)
f1_score_euclidean = f1_score(train_data_y, predicted_hierarchical_euclidean)

## Model and fit the training data to the AgglomerativeClustering model
## complete linkage + manhattan
manhattan_dst = pairwise_distances(pdm_train, metric='manhattan')
hierarchical_manhattan = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='complete')
predicted_hierarchical_manhattan = hierarchical_manhattan.fit_predict(manhattan_dst)
f1_score_manhattan = f1_score(train_data_y, predicted_hierarchical_manhattan)

## Model and fit the training data to the AgglomerativeClustering model
## average linkage + cosine
cosine_dst_avg = pairwise_distances(pdm_train, metric='cosine')
hierarchical_cosine_avg = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='average')
predicted_hierarchical_cosine_avg = hierarchical_cosine_avg.fit_predict(cosine_dst_avg)
f1_score_cosine_avg = f1_score(train_data_y, predicted_hierarchical_cosine_avg)


## Model and fit the training data to the AgglomerativeClustering model
## average linkage + euclidean
euclidean_dst_avg = pairwise_distances(pdm_train, metric='euclidean')
hierarchical_euclidean_avg = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='average')
predicted_hierarchical_euclidean_avg = hierarchical_euclidean_avg.fit_predict(euclidean_dst_avg)
f1_score_euclidean_avg = f1_score(train_data_y, predicted_hierarchical_euclidean_avg)


## Model and fit the training data to the AgglomerativeClustering model
## average linkage + manhattan
manhattan_dst_avg = pairwise_distances(pdm_train, metric='manhattan')
hierarchical_manhattan_avg = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='average')
predicted_hierarchical_manhattan_avg = hierarchical_manhattan_avg.fit_predict(manhattan_dst_avg)
f1_score_manhattan_avg = f1_score(train_data_y, predicted_hierarchical_manhattan_avg)

print("F1-score for complete linkage + cosine", f1_score_cosine)
print("F1-score for complete linkage + euclidean", f1_score_euclidean)
print("F1-score for complete linkage + manhattan", f1_score_manhattan)
print("F1-score for average linkage + cosine", f1_score_cosine_avg )
print("F1-score for average linkage + euclidean", f1_score_euclidean_avg)
print("F1-score for average linkage + manhattan", f1_score_manhattan_avg)

#Agglomerative clustering uses a bottom-up approach, wherein each data point starts in its own cluster. These clusters are then joined, by taking the two most similar clusters together and merging them.
#In all the above functions, AgglomerativeClustering function has parameters, n_clusters means the number of clusters to find, affinity is used to find the distance, If “precomputed”, a distance matrix is needed as input for the fit method. , linkage= The linkage criterion determines which distance to use
print(" The best Hierarchical Agglomerative Clustering Model: model that is built using average linkage and the one that uses manhattan distance. This model has a score 0.3980952380952381 which is higher than thae rest." )


###################end code for Task 2-a
```
